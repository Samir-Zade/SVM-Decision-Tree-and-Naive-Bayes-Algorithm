# **Support Vector Classifier**

```python
#import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#let's create the data points
from sklearn.datasets import make_classification
```

```python
#define the dependent and independent variable
X, y = make_classification(n_samples = 1000, n_features = 2,
                           n_classes = 2, n_clusters_per_class = 2, n_redundant = 0)
```

```python
X
```

```python
y
```

```python
#visualze the data
sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue = y)
plt.show()
```

```python
#define the train and test data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 47)
```

```python
#import the support vector classifier
from sklearn.svm import SVC

#define the model
svc = SVC(kernel = "linear")
```

```python
#fit the model
svc.fit(X_train, y_train)
```

```python
svc.coef_
```

```python
#making the prediction with model
y_pred = svc.predict(X_test)
y_pred
```

```python
#finding the accuracy, confusion matrix and classification report
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

#printing the metrics
print(accuracy_score(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

## Hyper Parameter Tuning with SVC

```python
#import the required libraries
from sklearn.model_selection import GridSearchCV

#define the parameters
parameters = {
    "C" : [0.1, 1, 10, 100, 1000],
    "gamma" : [1, 0.1, 0.01, 0.001, 0.0001]
}
```

```python
#define the grid search cv
grid = GridSearchCV(svc, param_grid = parameters, cv = 5, refit = True, verbose = 3)

#fit the model
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
#predicting the model
y_pred2 = grid.predict(X_test)

#printing the metrics
print(accuracy_score(y_pred2, y_test))
print(confusion_matrix(y_pred2, y_test))
print(classification_report(y_pred2, y_test))
```

Hyper Parameter Tuning with SVC **with kernel = 'linear'**

```python
#import the required libraries
from sklearn.model_selection import GridSearchCV

#define the parameters
parameters = {
    "C" : [0.1, 1, 10, 100, 1000],
    "gamma" : [1, 0.1, 0.01, 0.001, 0.0001],
    "kernel" : ["linear"]
}
```

```python
#define the grid search cv
grid = GridSearchCV(svc, param_grid = parameters, cv = 5, refit = True, verbose = 3)

#fit the model
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
#predicting the model
y_pred3 = grid.predict(X_test)

#printing the metrics
print(accuracy_score(y_pred3, y_test))
print(confusion_matrix(y_pred3, y_test))
print(classification_report(y_pred3, y_test))
```
